```python
import pandas as pd
import numpy as np
import datetime
import re
import matplotlib.pyplot as plot
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
```

### Importamos bibliotecas que quizás usaremos y el nuevo DataFrame previamente creado.

```python
shark = pd.read_csv("../data/sharkfinallimpio.csv",index_col=0)
shark = pd.DataFrame(shark)

shark.head()
```

### Para ponernos en situación he creado unos graficos generales: 

#### El primero de ellos nos muestra la cantidad de ataques de tiburón para los deportes de la hipotesis: SURF, NADO y PESCA.
#### Se observa una progresion ascendente , quizás por el ascendo de la practica de estos deportes. 

#### En el segundo se puede ver de una manera muy rápida la cantidad de ataques de tiburón para los paises seleccionados: USA, AUS, ShOUT AFRICA, PAPUA NUEVA GUINEA y NUEVa ZELANDA.


#### Continuamos mostrando los ataques por cada deporte de manera individual.

#### Y el 4º grafico es una combinación del 2º y 3º. 






```python
fig, ax = plot.subplots(2, 2)
fig.set_size_inches(16, 9)
palette = sns.color_palette("cubehelix")

shark["Year"].value_counts().sort_index().plot(ax=ax[0,0], color = "slategrey")
shark.dead.value_counts().plot(kind="bar", ax = ax [0,1], color = "slategrey")

sns.countplot(x = shark.Activity, palette = palette, ax = ax[1,0])
sns.countplot(y = shark.Activity, hue = shark.dead, palette= palette, ax= ax[1,1])

ax[0, 0].set_title("Cantidad ataque tiburones en el tiempo")
ax[0, 1].set_title("Mortalidad de los ataques")
ax[1, 0].set_title("Deportes con más atauqes")
ax[1, 1].set_title("Mortalidad en los deportes")
```

##### Con este grafico podemos deducir que la mayor parte de los ataques son a hombres.

```python
figura2 = sns.displot(x = shark.Year, hue= shark.Sex, col = shark.Activity,  binwidth=20 , color = "darkkhaki")
plot.figure(figsize=(15,25))
```
