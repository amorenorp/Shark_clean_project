```python
import pandas as pd
import numpy as np
import datetime
import re
import matplotlib.pyplot as plot
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
```

```python
shark = pd.read_csv("../data/sharkfinallimpio.csv",index_col=0)
shark = pd.DataFrame(shark)

shark.head()
```

```python
fig, ax = plot.subplots(2, 2)
fig.set_size_inches(16, 14)

shark["Year"].value_counts().sort_index().plot(ax=ax[0,0])
shark.dead.value_counts().plot(kind="bar", ax = ax [0,1])

sns.countplot(x = shark.Activity, palette = "Set2", ax = ax[1,0])
sns.countplot(y = shark.Activity, hue = shark.dead, palette= "Set2", ax= ax[1,1])


```

```python
sns.displot(x = shark.Country, multiple="dodge", col= shark.Activity)
```

```python
sns.displot(x = shark.Country, multiple="dodge", col= shark.dead)
```
